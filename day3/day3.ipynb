```python
from string import ascii_lowercase
from string import ascii_uppercase

priorities = {}
priority = 1
for c in ascii_lowercase:
    priorities[c] = priority
    priority+=1
    
for c in ascii_uppercase:
    priorities[c] = priority
    priority+=1

print(priorities)
```

```python
with open('day3.txt') as f:
    lst = f.readlines()
    tesf = [x.strip() for x in lst]
    
str1 = 'vJrwpWtwJgWr'
str2 = 'hcsFMMfFFhFp'


''.join(set(str1).intersection(str2))
```

```python
letters = []

for word in tesf:
    firstpart, secondpart = word[:len(word)//2], word[len(word)//2:]
    letter = ''.join(set(firstpart).intersection(secondpart))
    letters.append(letter)
    
letters
```

```python
prios = [priorities[k] for k in letters]
prios
```

```python
sum(prios)
```

```python
str1 = 'vJrwpWtwJgWrhcsFMMfFFhFp'
str2 = 'jqHRNqRjqzjGDLGLrsFMfFZSrLrFZsSL'
str3 = 'PmmdzqPrVvPwwTWBwg'

''.join(set(str1).intersection(str2).intersection(str3))
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
        
new_groups = list(chunks(tesf, 3))
```

```python

new_prio = []
for group in new_groups:
    str1, str2, str3 = group[0], group[1], group[2]
    x = ''.join(set(str1).intersection(str2).intersection(str3))
    new_prio.append(x)
    
new_prio
```

```python
prios_new = [priorities[k] for k in new_prio]
sum(prios_new)
```
